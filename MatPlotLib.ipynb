```python
import numpy as np
x = np.linspace(0,2*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize = (8,6))
ax1.plot(x,y_sin, color = 'purple',label= 'sin(x)',linewidth=2)
ax1.set_title("Sine Function")
ax1.set_ylabel("Amplitude")
ax1.legend()
ax1.grid(True)

ax2.plot(x,y_cos, color = 'teal',label= 'cos(x)',linewidth=2,linestyle="--")
ax2.set_title("Cosine Function")
ax2.set_xlabel("Angle(radians)")
ax2.set_ylabel("Amplitude")
ax2.legend()
ax2.grid(True)

ax3.plot(x,y_tan, color = 'green',label= 'tan(x)',linewidth=2,linestyle="--")
ax3.set_title("Tan Function")
ax3.set_xlabel("Angle(radians)")
ax3.set_ylabel("Amplitude")
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 🎓 Dummy user data for one subject (e.g., AI)
users = ["Diwakar", "Amit", "Sneha", "Ravi", "Tina"]
avg_scores = [85, 60, 45, 72, 90]  # X-axis
attempts = [5, 2, 4, 3, 6]         # Y-axis

# 🌈 Assign color based on score
colors = []
for score in avg_scores:
    if score >= 80:
        colors.append("green")   # High
    elif score >= 60:
        colors.append("orange")  # Medium
    else:
        colors.append("red")     # Low

# 📊 Plot
plt.figure(figsize=(10, 6))
plt.scatter(avg_scores, attempts, c=colors, s=100, edgecolors='black')

# 🏷️ Labels & title
for i, user in enumerate(users):
    plt.text(avg_scores[i] + 0.5, attempts[i] + 0.1, user, fontsize=9)

plt.title("User Performance Scatter Plot (Subject: AI)", fontsize=14)
plt.xlabel("Average Score (%)", fontsize=12)
plt.ylabel("Quiz Attempts", fontsize=12)
plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, max(attempts) + 2)

# 🟢 Legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='green', edgecolor='black', label='High (80%+)'),
    Patch(facecolor='orange', edgecolor='black', label='Medium (60–79%)'),
    Patch(facecolor='red', edgecolor='black', label='Low (<60%)')
]
plt.legend(handles=legend_elements, title="Performance", loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the pie charts
categories = ['Correct', 'Incorrect']
colors = ['#4CAF50', '#F44336']

# Performance data
easy_data = [75, 25]
medium_data = [60, 40] 
hard_data = [45, 55]

# Create figure
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Machine Learning Performance Level Wise', fontsize=16, fontweight='bold')

# Titles for individual pie charts
titles = ['Easy', 'Medium', 'Hard']
data_list = [easy_data, medium_data, hard_data]

# Create all three pie charts in a loop
for i, (ax, title, data) in enumerate(zip(axes, titles, data_list)):
    wedges, texts, autotexts = ax.pie(data, autopct='%1.1f%%', colors=colors,
                                     startangle=90, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
    ax.set_title(title, fontsize=14, fontweight='bold')
    
    # Add labels only to the first chart to avoid repetition
    if i == 0:
        ax.legend(wedges, categories, loc='upper left', bbox_to_anchor=(-0.1, -0.1))

plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
